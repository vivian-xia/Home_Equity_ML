## Assignment Data Normalize/Standardize

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import MinMaxScaler #range between 0 to 1
from sklearn.preprocessing import StandardScaler #range between +-3
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
FILE   = "HMEQ_Loss.csv"

df = pd.read_csv( FILE )
df.head()
```

```python
TARGET_FLAG = 'TARGET_BAD_FLAG'
TARGET_LOSS = 'TARGET_LOSS_AMT'
```

```python
dt = df.dtypes

#create a new list of categorical and numerical variables
dt = df.dtypes
objList = []
numList = []
for i in dt.index :
    if i in ( [ TARGET_FLAG, TARGET_LOSS ] ) : continue
    if dt[i] in (["object"]) : objList.append( i )
    if dt[i] in (["float64","int64"]) : numList.append( i )


print(" OBJECTS ")
print(" ------- ")
for i in objList :
    print( i )
print(" ------- \n\n")


print(" NUMBER ")
print(" ------- ")
for i in numList :
    print( i )
print(" ------- ")
```

```python
X = df.copy() 
X = X.drop(objList, axis=1 ) #all numeric values
X = X.drop([TARGET_FLAG,TARGET_LOSS], axis=1 )
varNames = X.columns #put column names into the list

print( X.head().T )
print( "\n\n")
print( X.describe() )
```

### NORMALIZE THE DATA

```python
theScaler = MinMaxScaler()
theScaler.fit( X )

X_MINMAX = theScaler.transform( X ) #transform new data
X_MINMAX = pd.DataFrame( X_MINMAX )
print( X_MINMAX.head() )
```

```python
varNames_minmax = [] #make a new list
for i in varNames:
  newName = "nor_" + i
  varNames_minmax.append(newName)
print(varNames_minmax)
print("\n\n")

X_MINMAX.columns = varNames_minmax
print(X_MINMAX.head())
print("\n\n")
print(X_MINMAX.describe()) #min and max should always be 0 and 1 respectively
print("\n\n")
```

```python
X_MINMAX[ "TARGET_BAD_FLAG" ] = df['TARGET_BAD_FLAG']
X_MINMAX[ "TARGET_LOSS_AMT" ] = df['TARGET_LOSS_AMT']
for i in objList:
    X_MINMAX[i] = df[i]

print( X_MINMAX.head() )
```

```python
X_NEW = pd.concat([ X , X_MINMAX ], axis=1 )
print( X_NEW.head() )
```

#### Remove outliers and rerun

```python
df.describe().T
```

```python
dt = df.dtypes
numList = []
for i in dt.index :
    if i in ( [ TARGET_FLAG, TARGET_LOSS ] ) : continue
    if dt[i] in (["float64","int64"]) : numList.append( i )
```

```python
for i in numList :
    print( i )
```

```python
truncList = []

for i in numList :
    theMean = df[i].mean()
    theSD = df[i].std()
    theMax = df[i].max()
    theCutoff = round( theMean + 3*theSD )
    if theMax < theCutoff : continue
    FLAG = "O_" + i
    TRUNC = "TRUNC_" + i
    df[ FLAG ] = ( df[i] > theCutoff )+ 0
    df[ TRUNC ] = df[ i ]
    truncList.append(TRUNC) #use list for numerical columns for data transformation
    df.loc[ df[TRUNC] > theCutoff, TRUNC ] = theCutoff
    df = df.drop( i, axis=1 )
```

```python
df.describe().T
```

#### Normalize Data

```python
X = df.copy() 
X = X[truncList]
varNames = X.columns 

print( X.head().T )
print( "\n\n")
print( X.describe() )
```

```python
theScaler = MinMaxScaler()
theScaler.fit( X )

X_MINMAX = theScaler.transform( X ) 
X_MINMAX = pd.DataFrame( X_MINMAX )
print( X_MINMAX.head() )
```

```python
varNames_minmax = []
for i in varNames:
  newName = "nor_" + i
  varNames_minmax.append(newName)
print(varNames_minmax)
print("\n\n")

X_MINMAX.columns = varNames_minmax
print(X_MINMAX.head())
print("\n\n")
print(X_MINMAX.describe()) #min and max should always be 0 and 1 respectively
print("\n\n")
```

```python
X_MINMAX[ "TARGET_BAD_FLAG" ] = df['TARGET_BAD_FLAG']
X_MINMAX[ "TARGET_LOSS_AMT" ] = df['TARGET_LOSS_AMT']
for i in objList:
    X_MINMAX[i] = df[i]

print( X_MINMAX.head() )
```

```python
X_NEW = pd.concat([ X , X_MINMAX ], axis=1 )
print( X_NEW.head() )
```
